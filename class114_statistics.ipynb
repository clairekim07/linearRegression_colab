```python
import csv
import pandas as panda
import plotly.express as px

df = panda.read_csv("main.csv")

GRE = df["GRE Score"].tolist()
chance = df["Chance of Admit "].tolist()

m = .011
b = -2.77
y = []

for x in GRE:
  yValue = m*x+b
  y.append(yValue)

fig = px.scatter(x=GRE, y=chance)
fig.update_layout(shapes=[dict(type='line',y0=min(y),y1=max(y),x0=min(GRE),x1=max(GRE))])
fig.show()

print(f"Chance of Admit is {GRE} for GRE Score of {y}")
```
